```
#importing some required libraries 

import numpy as np 

import pandas as pd 

import sklearn

from sklearn.model_selection import train_test_split

import sklearn.metrics as metrics

from sklearn.metrics import confusion_matrix 

from sklearn.tree import DecisionTreeClassifier 

from sklearn.metrics import accuracy_score 

from sklearn.datasets import load_iris

from matplotlib import pyplot as plt

from sklearn import datasets

from sklearn import tree
```

```
#Loading datasets 

iris_data = load_iris() 

iris=pd.DataFrame(iris_data.data)

#priting features name of iris data 

print ("Features Name : ", iris_data.feature_names) 

#shape of datasets 

print ("Dataset Shape: ", iris.shape) 

#first five sample 

print ("Dataset: ",iris.head())  
```

```
#priting samples and target 

X = iris.values[:, 0:4] 

Y = iris_data.target

print(X)

print(Y)
```

```
iris["types"] = Y
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
sns.set(font_scale=1.5)
sns.pairplot(iris,hue="types",size=3);
plt.show()    #relationship between the features
```

```
# Splitting the dataset into train and test 

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```
# Decision tree classifier 

clf= DecisionTreeClassifier(random_state = 100)

#fitting the training data

clf.fit(X_train, y_train)
```

```
# prediction on random data

X=[[6.4,1.8 ,3.6 ,0.5]]

Y_pred=clf.predict(X)

print(Y_pred)

# prediction on X_test (testing data )

Y_pred=clf.predict(X_test)

print(Y_pred)
```

```
#Accuray of the model 
import sklearn.metrics as metrics
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))
from sklearn.metrics import  confusion_matrix

cm=np.array(confusion_matrix(y_test,Y_pred))

cm
```
